```python
# Visualizations (you can add more visualizations based on your specific analysis needs)
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of GHI (Global Hunger Index)
plt.figure(figsize=(8, 6))
sns.histplot(df['GHI'], bins=20, kde=True, color='blue')
plt.title('Distribution of Global Hunger Index (GHI)')
plt.xlabel('GHI')
plt.ylabel('Frequency')
plt.show()
```

```python
cols  = (df.columns.values)
firstset = cols[0:7]
secondset = cols[7:]
np.append(secondset, 'GHI')
sns.pairplot(df[firstset])
plt.show()

sns.pairplot(df[secondset])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram and Density Plot for a numerical variable
plt.figure(figsize=(8, 6))
sns.histplot(df['GHI'], bins=20, kde=True, color='blue')
plt.title('Histogram and Density Plot of Global Hunger Index (GHI)')
plt.xlabel('GHI')
plt.ylabel('Density')
plt.show()
```

```python
# Box Plot for a numerical variable
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Unemployment Above age 15'])
plt.title('Box Plot of Unemployment Above age 15')
plt.xlabel('Unemployment Above age 15')
plt.show()
```

```python
# Violin Plot for a numerical variable
plt.figure(figsize=(8, 6))
sns.violinplot(x=df['Unemployment Above age 15'])
plt.title('Violin Plot of Unemployment Above age 15')
plt.xlabel('Unemployment Above age 15')
plt.show()
```

```python
# Scatter Plot with Regression Line
plt.figure(figsize=(8, 6))
sns.regplot(x='Import (US$ Thousand)', y='Export (US$ Thousand)', data=df)
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Import (US$ Thousand)')
plt.ylabel('Export (US$ Thousand)')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Bar Plot for a categorical variable
plt.figure(figsize=(8, 6))
sns.countplot(x=df['Year'] // 10 * 10)  # Grouping years by decade
plt.title('Count of Observations by Decade')
plt.xlabel('Decade')
plt.ylabel('Count')
plt.show()
```
